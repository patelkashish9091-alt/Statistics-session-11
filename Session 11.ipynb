```python
import numpy as np
import matplotlib.pyplot as plt

# Skewed dataset (Instagram likes)
np.random.seed(42)
likes = np.random.exponential(scale=100, size=10000)

sample_means = []

# Take 1000 samples of size 30
for i in range(1000):
    sample = np.random.choice(likes, size=30)
    sample_means.append(np.mean(sample))

# Plot histogram of sample means
plt.hist(sample_means, bins=30, edgecolor='black')
plt.title("Central Limit Theorem (Sample Size = 30)")
plt.xlabel("Sample Mean")
plt.ylabel("Frequency")
plt.grid(True)

plt.show()
```

The Central Limit Theorem states that when we take many random samples from a population and calculate the mean of each sample, the distribution of those sample means becomes approximately normal, even if the original data is not normally distributed.

Example: Suppose Zomato order amounts are highly skewed because some customers place very expensive orders while most place smaller orders. If we repeatedly take random samples of 30 orders and calculate the average order amount for each sample, the averages will form a bell-shaped (normal) distribution. This helps businesses make accurate predictions and statistical analyses using sample data instead of the entire population.

Real-World Scenario (Flipkart)

The Central Limit Theorem is useful for analyzing the average delivery time of Flipkart orders. Instead of checking every order, the company can take many random samples of deliveries and calculate the average delivery time for each sample. According to the CLT, these sample averages will follow a normal distribution. This allows Flipkart to estimate overall delivery performance, identify delays, and make better logistics decisions without analyzing millions of records.

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)

# Skewed dataset
likes = np.random.exponential(scale=100, size=10000)

sample_means = []

# 1000 samples of size 5
for i in range(1000):
    sample = np.random.choice(likes, size=5)
    sample_means.append(np.mean(sample))

plt.hist(sample_means, bins=30, edgecolor='black')
plt.title("CLT Simulation (Sample Size = 5)")
plt.xlabel("Sample Mean")
plt.ylabel("Frequency")
plt.grid(True)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(50)

# Exponential (non-normal) population
population = np.random.exponential(scale=2, size=10000)

sample_means = []

# Generate 1000 samples
for i in range(1000):
    sample = np.random.choice(population, size=30)
    sample_means.append(np.mean(sample))

# Plot histogram
plt.hist(sample_means, bins=30, edgecolor='black')
plt.title("Central Limit Theorem using Exponential Distribution")
plt.xlabel("Sample Mean")
plt.ylabel("Frequency")
plt.grid(True)

plt.show()
```
